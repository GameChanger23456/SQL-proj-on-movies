```python
import sqlite3
import pandas as pd

# Load the CSV file
df = pd.read_csv("/kaggle/input/movies/movies_data.csv")

# Display first few rows
df.head()
```

```python
# Create a connection to a new SQLite database file
conn = sqlite3.connect('movies.db')

# Write the dataframe to a SQL table named 'movies'
df.to_sql('movies', conn, if_exists='replace', index=False)

# Verify table creation
pd.read_sql_query("SELECT name FROM sqlite_master WHERE type='table';", conn)
```

```python
pd.read_sql_query("SELECT * FROM movies LIMIT 5;", conn)
```

```python
pd.read_sql_query("SELECT name,genre FROM movies;", conn) #list of films titles genres
```

```python
pd.read_sql_query("""
SELECT name,rating
FROM movies
ORDER BY rating DESC
LIMIT 5;
""", conn) #top5 films based on ratings
```

```python
pd.read_sql_query("""
SELECT genre,ROUND(AVG(rating), 2) AS avg_rating
FROM movies
GROUP BY genre
ORDER BY avg_rating DESC;
""", conn) #average rated films list
```

```python
pd.read_sql_query("""
SELECT genre,COUNT(*) AS total_movies
FROM movies
GROUP BY genre
ORDER BY total_movies DESC;
""", conn) #films count genres wise
```

```python
pd.read_sql_query("""
SELECT name,year,rating
FROM movies
WHERE year > 2015
ORDER BY year;
""", conn) #films rel after 2015
```
